```python
import requests
from bs4 import BeautifulSoup
import re
import uuid
import csv
```

```python
headers = {"User-Agent": "insomnia/8.4.5"}
url = 'http://www.ufcstats.com/statistics/events/completed?page=all'
site = requests.get(url,headers=headers)
soup = BeautifulSoup(site.content, 'html.parser')
print(site)
```

```python
event_links = soup.find_all('td', class_=re.compile('b-statistics__table-col'))
```

```python
print(len(event_links))
```

```python
event_href = []
for event in event_links:
    try:
        event_href.append(event.find('a')['href'])
    except:
        pass
```

```python
event_href = event_href[1:]
```

```python
print(len(event_href))
```

```python
with open('event_UUID.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['UUID', 'name', 'date', 'location', 'link'])
```

```python
with open('fighter_links_UUID.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['link', 'UUID'])
```

```python
fighter_details_UUID = {}
for link in event_href:
    url = link
    headers = {"User-Agent": "insomnia/8.4.5"}
    site = requests.get(url,headers=headers)
    print(site)
    soup = BeautifulSoup(site.content, 'html.parser')
    event_name = soup.find('h2', class_=re.compile('b-content__title')).text.strip()
    details = soup.find_all('li', class_=re.compile('b-list__box-list-item'))
    date = details[0].text.strip()
    location = details[1].text.strip()
    date_clean = date.replace('\n', '').split(':')[1].lstrip()
    location_clean = location.replace('\n', '').split(':')[1].lstrip()
    entry = [uuid.uuid4(), event_name, date_clean, location_clean, link]
    with open('event_UUID.csv', 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(entry)
    
    fighters = soup.find_all('td', class_=re.compile('b-fight-details__table-col l-page_align_left'))
    for fight in fighters:
        iterate = fight.find_all('a', class_=re.compile('b-link b-link_style_black'))
        for item in iterate:
            item_href = item['href']
            if item_href not in fighter_details_UUID:
                new_uuid = uuid.uuid4()
                fighter_details_UUID[item_href] = new_uuid
                with open ("fighter_links_UUID.csv", 'a', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow([item_href, str(new_uuid)])
```
